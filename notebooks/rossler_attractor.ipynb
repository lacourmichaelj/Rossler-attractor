# Rössler Attractor Simulation & Visualization

This project simulates and visualizes the chaotic Rössler system using numerical ODE solvers in Python.

**Technologies:** NumPy, SciPy, Matplotlib

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Parameters
a = 0.1
b = 0.1
c = 18

# Fixed points
x = (c + np.sqrt(c**2 - 4*a*b)) / 2
y = (-c - np.sqrt(c**2 - 4*a*b)) / (2*a)
z = (c + np.sqrt(c**2 - 4*a*b)) / (2*a)

x1 = (c - np.sqrt(c**2 - 4*a*b)) / 2
y1 = (-c + np.sqrt(c**2 - 4*a*b)) / (2*a)
z1 = (c - np.sqrt(c**2 - 4*a*b)) / (2*a)

fixedpoint1 = [x, y, z]
fixedpoint2 = [x1, y1, z1]

# Rossler system
def ross(t, p):
    return [
        -p[1] - p[2],
        p[0] + a * p[1],
        b + (p[0] - c) * p[2]
    ]

# Time span
t0, tfinal = 0, 10

# Initial grid
x0 = np.arange(0, 21, 1)
y0 = np.arange(0, 21, 1)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

for k in range(len(y0)):
    sol = solve_ivp(ross, [t0, tfinal], [x0[k], y0[k], 1], t_eval=np.linspace(t0, tfinal, 500))
    ax.plot(sol.y[0], sol.y[1], sol.y[2])

# Plot fixed points
ax.scatter(x, y, z, color='red', marker='*')
ax.scatter(x1, y1, z1, color='blue', marker='*')

ax.set_title('Rossler Attractor')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```
